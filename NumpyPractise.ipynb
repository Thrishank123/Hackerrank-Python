```python
import pandas as pd

d = {'one' : pd.Series([1, 2, 3], index=['a', 'b', 'c']),
     'two' : pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])}

df = pd.DataFrame(d)

# Adding a new column to an existing DataFrame object with column label by
print ("Adding a new column by passing as Series:")
df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)

print ("Adding a new column using the existing columns in DataFrame:")
df['four']=df['one']+df['three']

print(df)
```

```python
import pandas as pd

# Create the initial DataFrames
df = pd.DataFrame([[1, 2], [3, 4]], columns=['a', 'b'])
df2 = pd.DataFrame([[5, 6], [7, 8]], columns=['a', 'b'])

# Concatenate df and df2
df = pd.concat([df, df2])

# Reset the index to have a continuous range of indices
df = df.reset_index(drop=True)

# Print the DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np

def adder(ele1, ele2):
    return ele1 + ele2

df = pd.DataFrame([
    [-0.64330483, -1.30091727,  0.2070746],
    [ 0.12093637, -0.48941621,  0.005298],
    [-0.01346436, -1.54098767,  0.68901864],
    [-0.08757182,  0.17791822,  0.26038605],
    [ 2.05664329, -0.95093109,  0.65359171]
], columns=['col1', 'col2', 'col3'])

# Apply the adder function using pipe
df = df.pipe(adder, 2)

print(df)
```
